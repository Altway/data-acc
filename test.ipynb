```python
import seaborn as sns
import pandas as pd
import datetime

df = pd.read_csv("/home/rems/Desktop/SharedFolder/laaa.csv")


df["temperature"] = df["temperature"].apply(lambda x:float(x.replace(',','.')))
df["humidity"] = df["humidity"].apply(lambda x:float(x[:-1].replace(',','.')))
df["time"] = df["time"].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
print(df.info())
print(df)
df.to_csv(r'/home/rems/Desktop/SharedFolder/cleaned.csv')
sns.set_theme(style="ticks")


sns.scatterplot(x="time", y="temperature",data=df)#, col="time", hue="time", data=df,
           #col_wrap=2, ci=None, palette="muted", height=4,
           #scatter_kws={"s": 50, "alpha": 1})
```
